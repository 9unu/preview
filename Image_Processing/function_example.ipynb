# 3.2 문자가 있는 바운더리 구하기

`바운더리 좌표`는 왼쪽 모서리부터 시계방향으로 나옴

```python
boundary=[]
for box in box_loc:
    top= int(min([box[i][1] for i in range(4)]))+3  # +-3은 바운딩 박스랑 실제 글자 사이의 간격 때문
    bottom=int(max([box[i][1] for i in range(4)]))-3
    boundary.append((top,bottom))
```

```python
boundary
```

## 3.3 target값 매핑

```python
data['char']=0
for pair in boundary:
    start, end = pair
    data.iloc[start:end+1]['char']=1
```
